```python
NOISE_DIR = "./noise/"
```

```python
def save_wb_per_layer(X, NOISE_DIR, N=10):
    for i in range(N):
        NOISE_I_DIR = NOISE_DIR + str(i)
        os.makedirs(NOISE_I_DIR, exist_ok=True)
        Xg = noise_a(X, 0.2)
        Xs = noise_p(X, 64)
        Xm = noise_m(X, 15)
        Xo = noise_v(X, 192)
        np.save(os.path.join(NOISE_I_DIR, "gauss.npy"), Xg, allow_pickle=True)
        np.save(os.path.join(NOISE_I_DIR, "salt.npy"), Xs, allow_pickle=True)
        np.save(os.path.join(NOISE_I_DIR, "move.npy"), Xm, allow_pickle=True)
        np.save(os.path.join(NOISE_I_DIR, "occ.npy"), Xo, allow_pickle=True)
    print("Saved all noise data into separate files.")
```

```python
save_wb_per_layer(x_train, NOISE_DIR)
```
